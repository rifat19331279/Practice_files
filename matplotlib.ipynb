```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(5,100,100)
y=np.linspace(10,1000,100)
```

```python
plt.plot(x,y)
```

```python
plt.plot([1,4,6,8],[3,8,3,5])
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('current')
plt.ylabel('voltage')
plt.title('Ohm\'s law')

plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
x=np.linspace(0,10,20)
y=x*2

plt.plot(x,y,'r+')

plt.xlabel('current')
plt.ylabel('voltage')
plt.title('ohm\'s law')

plt.show()
```

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)

plt.plot(x,y,'r-',x,y**2,'b+',x,y**3,'g^')
plt.show()
```

```python
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title('y=logx')
plt.plot(x,y)

plt.subplot(122)
plt.title('y=logx**2')
plt.plot(x,y**2)
plt.show()
```

```python
x=np.linspace(1,10,100)

plt.figure(1)

plt.subplot(2,2,1)
plt.title('Linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('Cubic')
plt.plot(x,x**3)

plt.figure(2)

plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,np.exp(x))

plt.show()
```

```python
x=np.linspace(1,10,50)

plt.figure(1)

plt.subplot(3,4,1)
plt.title('linear')
plt.plot(x,x)
plt.show()

plt.subplot(3,4,2)
plt.title('sqr')
plt.plot(x,x**2)
plt.show()

plt.subplot(3,4,3)
plt.title('cub')
plt.plot(x,x**3)
plt.show()

plt.subplot(3,4,4)
plt.title('4th power')
plt.plot(x,x**4)
plt.show()

plt.figure(2)

plt.subplot(3,4,1)
plt.title('logx')
plt.plot(x,np.log(x))
plt.show()

plt.subplot(3,4,2)
plt.title('logx**2')
plt.plot(x,np.log(x)**2)
plt.show()

plt.subplot(3,4,3)
plt.title('logx**3')
plt.plot(x,np.log(x)**3)
plt.show()

plt.subplot(3,4,4)
plt.title('logx**4')
plt.plot(x,np.log(x)**4)
plt.show()

plt.figure(3)

plt.subplot(3,4,1)
plt.title('exp')
plt.plot(x,np.exp(x))
plt.show()

plt.subplot(3,4,2)
plt.title('exp**2')
plt.plot(x,np.exp(x)**2)
plt.show()

plt.subplot(3,4,3)
plt.title('exp**3')
plt.plot(x,np.exp(x)**3)
plt.show()

plt.subplot(3,4,4)
plt.title('exp**4')
plt.plot(x,np.exp(x)**4)
plt.show()
```
